### For Loop 

```python
seq= [1,2,3,4,5]
```

```python
for num in seq:
    print(num)
```

```python
for num in seq:
    print('Hello')
```

### While Loop 

```python
i=1
for i in seq:
    print('The square of {} is {}'.format(i,i**2))
    i=i+1
```

### Range 

```python
for x in range(0,5):
    print(x)
```

### List Comprehension

```python
x=[1,2,3,4,5,6,7,8]
out=[]
for i in x:
    out.append(i**2)
print(out)
```

```python
out=[i**2 for i in x]
out
```

```python
[i**3 for i in x]
```
