### Map Function 

```python
def times2(x):
    return(x*2)
times2(39)
```

```python
seq=list(range(1,20))
map(times2,seq)
```

```python
list(map(times2,seq))
```

### Lambda Expression 

```python
anonymous_function=lambda x:x*2
```

```python
anonymous_function(87)
```

```python
seq=list(range(1,20))
list(map(lambda x:x*2, seq))
```

### Filter Function 

```python
filter(lambda x:x%2==0,seq)
```

```python
list(filter(lambda x:x%2==0,seq))
```

```python
list(filter(lambda x:x%3==0,seq))
```

### Tuple Unpacking 

```python
x=[(1,2),(3,4),(5,6)]
for i in x:
    print(i)
```

```python
for (a,b) in x:
    print(a)
```

```python
for a,b in x:
    print(a)
    print(b)
```
